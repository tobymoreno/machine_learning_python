```python
def decay(l, n, v):
    if n<1:
        return
    else:
        nv = v / 2
#         print(n, nv)
        l.append(nv)
        decay(l, n - 1, nv)

def sigmoid(z):
    return 1 / (1 + math.pow(math.e, -(z)))

def decorate(element):
    date = element[0]
    open_ = element[1]
    high = element[2]
    low = element[3]
    close = element[4]
    return {"date" : date, "open" : open_, "high" : high, "close" : close, "low" : low }

def calculateMP(c):
    LW = max(c["high"] - max(c["open"], c["close"]), 0)
    body = max(c["open"], c["close"]) - min(c["open"], c["close"])
    SP = LW / (LW + body)


    SW = max(min(c["open"], c["close"]) - c["low"], 0)
    BP = SW / (SW + body)
    
    c["SP"] = SP
    c["BP"] = BP
    

def triangle_score(p2, p1, idx2, idx1):
    return (1 / (0.001 + abs(p2 - p1)) * (1 / abs(idx2 - idx1)))
```

```python
def PRESSURE_LIST(JD):
    DECBs = []

    for jrd in JD:
    #     print(jrd)
        decB = decorate(jrd)
        calculateMP(decB)
        DECBs.append(decB)
    #     print(decB)

    # change the weight list formula 1 / x
    WEIGHTED_LIST = [((l[i[0]]) * i[1]['SP'], l[i[0]] * i[1]['BP'], i[1]['high'], i[1]['low'], i[1]['date'])  for i in enumerate(DECBs)]


    SPLIST = []
    BPLIST = []
    sp_high = 0
    bp_low = 10000000.0

    for j in enumerate(WEIGHTED_LIST):
        k = j[0]
        i = j[1]
        d = j[1][4]
    #     print(i[0], i[1])
        if(i[2] > sp_high):
            sp_high = i[2]
            SPLIST.append((k,sp_high, i[0], d))

        if(i[3] < bp_low):
            bp_low = i[3]
            BPLIST.append((k,bp_low, i[1], d))
            
    return (SPLIST, BPLIST)
```

```python
def sentiment_strength(OPEN, HIGH, LOW, CLOSE):
    body = CLOSE - OPEN
    LENGTH = abs(HIGH - LOW)
    if body > 0:
        score = abs(body) / LENGTH
    else:
        score = -1 * abs(body) / LENGTH
    return score * .01
        
```

```python
def sentiment_score(JRDATA):


    # print(SPLIST)
    # print(BPLIST)
    
    SPLIST, BPLIST = PRESSURE_LIST(JRDATA)

    LONG = sum([i[2] for i in SPLIST])
    SHORT = sum([i[2] for i in BPLIST])

#     print("SELLING PRESSURE=", LONG)
#     for i in SPLIST:
#         print(i[0], i[2], i[3])

#     print("BUYING PRESSURE=", SHORT)
#     for i in BPLIST:
#         print(i[0], i[2], i[3])

    PRESSURE_SCORE = sigmoid(LONG - (SHORT))
#     print(PRESSURE_SCORE)

    SPTLIST = []
    RSPLIST = [i for i in reversed(SPLIST)]
    for i in RSPLIST:
        # we want to find the next highest first
        nhigh = 0
        index = -1
        for j in RSPLIST:
            if(i[0] != j[0]):
                if j[1] > nhigh and j[1] < i[1]:
                    nhigh = j[1]
                    index = j[0]
    #                 print((i[3], i[1]), (j[3], j[1]))

        if(index > 0):
#             print(i[0], i[3], i[1], nhigh, index, i[0] - index)
            ts = triangle_score(i[1], nhigh, i[0], index)
    #         print(ts)
            SPTLIST.append(ts)
    SPTSCORE = sum(SPTLIST)
#     print("TRITOP", SPTSCORE)
    BPTLIST = []
    RBPLIST = [i for i in reversed(BPLIST)]
    for i in RBPLIST:
        nlow = 1000000
        index = -1    
        for j in RBPLIST:
            if(i[0] != j[0]):
    #             print((i[3], i[1]), (j[3], j[1]))
    #             print("x", i[1], j[1])
                if(j[1] < nlow and i[1] < j[1]):
                    nlow = j[1]
                    index = j[0]
    #     print(i, nlow, index)

        if(index > 0):
#             print(i[0], i[3], i[1], nlow, index, i[0] - index)
            ts = triangle_score(i[1], nlow, i[0], index)
    #         print(ts)
            BPTLIST.append(ts)

#     print(BPTLIST)
    BPTSCORE = sum(BPTLIST)    
#     print("TRIBOTTOM", BPTSCORE)
    TRI_SCORE = sigmoid(SPTSCORE - BPTSCORE)
#     print(TRI_SCORE)


    OVERALL_SCORE = np.average([PRESSURE_SCORE, TRI_SCORE])
#     print(JRDATA[0][0], OVERALL_SCORE)

    OPEN = JRDATA[0][1]
    HIGH = JRDATA[0][2]
    LOW = JRDATA[0][3]
    CLOSE = JRDATA[0][4]

    S_SCORE = sentiment_strength(OPEN, HIGH, LOW, CLOSE)
#     print("S", S_SCORE)


    OVERALL_SCORE = OVERALL_SCORE + S_SCORE
    return OVERALL_SCORE
```
