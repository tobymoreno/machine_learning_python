```python
def decay(l, n, v):
    if n<1:
        return
    else:
        nv = v / 2
        print(n, nv)
        l.append(nv)
        decay(l, n - 1, nv)

def sigmoid(z):
    return 1 / (1 + math.pow(math.e, -(z)))

def decorate(element):
    date = element[0]
    open_ = element[1]
    high = element[2]
    low = element[3]
    close = element[4]
    return {"date" : date, "open" : open_, "high" : high, "close" : close, "low" : low }

def calculateMP(c):
    LW = max(c["high"] - max(c["open"], c["close"]), 0)
    body = max(c["open"], c["close"]) - min(c["open"], c["close"])
    SP = LW / (LW + body)


    SW = max(min(c["open"], c["close"]) - c["low"], 0)
    BP = SW / (SW + body)
    
    c["SP"] = SP
    c["BP"] = BP
    

def triangle_score(p2, p1, idx2, idx1):
    return (1 / (0.001 + abs(p2 - p1)) * (1 / abs(idx2 - idx1)))
```

```python
def sentiment_strength(OPEN, HIGH, LOW, CLOSE):
    body = CLOSE - OPEN
    LENGTH = abs(HIGH - LOW)
    if body > 0:
        score = abs(body) / LENGTH
    else:
        score = -1 * abs(body) / LENGTH
    return score * .01
        
```
